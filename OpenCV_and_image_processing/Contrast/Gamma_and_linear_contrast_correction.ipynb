```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
# os.chdir(r"D:\NLP 1\OpenCV_tutorial\My_assignments\Exc_4")
```

```python
def gamma_correction(image, gamma):
    if type(image) == str:
        image = cv2.imread(image)
    lookUpTable = np.empty((1,256), np.uint8)

    for i in range(256):
        lookUpTable[0,i] = np.clip(pow(i / 255.0, gamma) * 255.0, 0, 255)
    gamma_corrected_image = cv2.LUT(image, lookUpTable)

    return gamma_corrected_image
```

```python
img_dir = "original_image.jpg"
original_image = cv2.imread(img_dir)
```

```python
alpha = 0.7
beta = 100
linear_contrast_ehnancement = cv2.convertScaleAbs(original_image, alpha=alpha, beta=beta)

plt.figure(figsize=[10,5])
plt.subplot(121);plt.imshow(original_image[...,::-1]);plt.title("Original");
plt.subplot(122);plt.imshow(linear_contrast_ehnancement[...,::-1]);plt.title("Linear contrast ehnancement");
```

```python
gamma = 0.25
gamma_corrected_image = gamma_correction(image=original_image, gamma=gamma)

plt.figure(figsize=[10,5])
plt.subplot(121);plt.imshow(original_image[...,::-1]);plt.title("Original");
plt.subplot(122);plt.imshow(gamma_corrected_image[...,::-1]);plt.title("Gamma corrected");
```

```python
cv2.imwrite("gamma_corrected_image.png", gamma_corrected_image)
cv2.imwrite("linear_contrast_ehnancement.png", linear_contrast_ehnancement)
```
